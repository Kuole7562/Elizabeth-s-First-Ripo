Part 2: Recreating Share of the population that is undernourished Data Visualization.

```python
#Import the necessary tools
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_undernorurishment=pd.read_csv('/content/prevalence-of-undernourishment 1.csv')
df_undernorurishment.head()
```

```python
# Rename the column to something more manageable since the long name, 2.1.1 Prevalence of undernourishment, raises a key error.
df_undernorurishment.rename(columns={df_undernorurishment.columns[-1]: "Undernourishment"}, inplace=True)

#Check if it worked
print(df_undernorurishment.head())
```

```python
# Convert 'Year' to integer and 'Undernourishment (%)' to float
df_undernorurishment["Year"] = df_undernorurishment["Year"].astype(int)
df_undernorurishment["Undernourishment"] = df_undernorurishment["Undernourishment"].astype(float)

# Filter data for specific regions or the world
selected_entities = ["Sub-Saharan Africa (FAO)","Southern Asia (FAO)", "World", "Northern Africa (FAO)", "South America (FAO)","South-eastern Asia (FAO)","Central Asia (FAO)"]
df_filtered = df_undernorurishment[df_undernorurishment["Entity"].isin(selected_entities)]

# Plot the data
plt.figure(figsize=(10,5))
for entity in selected_entities:
    subset = df_filtered[df_filtered["Entity"] == entity]
    plt.plot(subset["Year"], subset["Undernourishment"], marker="o", label=entity)

# Labels, Title, and Legend
plt.xlabel("Year")
plt.ylabel("Prevalence of Undernourishment (%)")
plt.title("Share of the population that is undernourished")
plt.legend()
plt.grid(True)
plt.show()
```
